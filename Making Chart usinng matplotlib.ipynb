```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
year = list(range(6))
india_gdp = [10,7,6,7,3,6]
china_gdp = [11,8,3,1,1,5]
```

```python
plt.plot(year, india_gdp, label = "India", linewidth= '3')
plt.plot(year, china_gdp, label = "China")
style.use('ggplot')
plt.title("GDP by COUNTRY")
plt.xlabel("year")
plt.ylabel("%change")
plt.legend()
plt.show()
```

```python
plt.hist(india_gdp, label = "india gdp")
plt.legend()
plt.show()
```
