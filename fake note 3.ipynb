```python
import pandas as pd
```

```python
data = pd.read_csv("files/note.csv")
```

```python
data
```

```python
import matplotlib.pyplot as plt
```

```python
data["v11"] = (data["v1"] - min(data["v1"])) / (max(data["v1"]) - min(data["v1"]))
data["v22"] = (data["v2"] - min(data["v2"])) / (max(data["v2"]) - min(data["v2"]))

data["v22"]
```

```python
from sklearn.cluster import KMeans
```

```python
import numpy as np
np_data = np.column_stack((data['v11'], data['v22']))
grouping = KMeans(n_clusters = 10).fit(np_data)
```

```python
cluster_points = grouping.cluster_centers_
cluster_points
```

```python
np_data
```
