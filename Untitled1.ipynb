```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("happy.csv")
```

```python
income = data['avg_income']
happyness = data['happyScore']
ineq = data["income_inequality"]
country = data["country"]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(income, happyness, s=ineq, alpha=0.25)
plt.xlabel("income")
plt.ylabel("happyness")
plt.title("HAPPYNESS INDEX BY COUNTRY")
for i in range(0,len(income),7):
    plt.text(income[i], happyness[i], country[i], color="red", c= "orange")
```

* chosing income, happyness and inequlity is looking more corelated to each other
* shorted the data using pandas functionality, get_sorted()
* label points are providng overview of data points over dataframe
