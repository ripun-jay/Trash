```python
import pandas as pd
```

```python
data = pd.read_csv("files/note.csv")
```

```python
data
```

```python
train_data = data.loc[100:1300]
```

```python
train_data
```

```python
test_data1 = data.loc[0:100]  #(data.loc[1300:1372]) 
```

```python
test_data1
```

```python
test_data2 = data.loc[1300:]
```

```python
test_data2
```

```python
test_data = pd.merge(test_data1, test_data2,how = "outer")

test_data
```

```python
train_data_x = train_data.drop("c", axis=1)
```

```python
train_data_x
```

```python
train_data_y = train_data["c"]
```

```python
train_data_y
```

```python
test_data_x = test_data.drop("c", axis=1)
test_data_x
```

```python
test_data_y = test_data["c"]
test_data_y
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier()
```

```python
clf.fit(train_data_x,train_data_y)
```

```python
clf.predict(train_data_x)
```

```python
predicted = clf.predict(train_data_x)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.scatter(np.arange(1201), predicted)
plt.scatter(np.arange(1201), 1.5*train_data_y, c= "green")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
error = mean_absolute_error(train_data_y, predicted)
error
```

```python
error2 = mean_squared_error(train_data_y, predicted)
```
